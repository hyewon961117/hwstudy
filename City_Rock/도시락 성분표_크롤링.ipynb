```python
import requests
from bs4 import BeautifulSoup
import time
from tqdm.notebook import tqdm
from urllib.parse import quote

import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
url = 'https://www.fatsecret.kr/%EC%B9%BC%EB%A1%9C%EB%A6%AC-%EC%98%81%EC%96%91%EC%86%8C/%EC%9D%BC%EB%B0%98%EB%AA%85/%EB%8B%A4%EC%9D%B4%EC%96%B4%ED%8A%B8-%EB%8F%84%EC%8B%9C%EB%9D%BD'
soup = BeautifulSoup(requests.get(url).text, 'html.parser')
soup
```

```python
div_list = soup.select('div.nutrition_facts.international div')

열량 = "-"
칼로리 = "-"
탄수화물 = "-"
설탕당 = "-"
단백질 = "-"
지방 = "-"
포화지방 = "-"
트랜스지방 = "-"
콜레스테롤 = "-"
나트륨 = "-"

for i in range(len(div_list)):
    div = div_list[i]
    
    if div.get_text()=="열량":
        열량 = div_list[i+1].get_text()

    elif "kcal" in div.get_text():
        칼로리 = div.get_text()

    elif div.get_text()=="탄수화물":
        탄수화물 = div_list[i+1].get_text()
    
    elif div.get_text()=="설탕당":
        설탕당 = div_list[i+1].get_text()
        
    elif div.get_text()=="단백질":
        단백질 = div_list[i+1].get_text()
    
    elif div.get_text()=="지방":
        지방 = div_list[i+1].get_text()
        
    elif div.get_text()=="포화지방":
        포화지방 = div_list[i+1].get_text()
    
    elif div.get_text()=="트랜스 지방":
        트랜스지방 = div_list[i+1].get_text()
    
    elif div.get_text()=="콜레스테롤":
        콜레스테롤 = div_list[i+1].get_text()
    
    elif div.get_text()=="나트륨":
        나트륨 = div_list[i+1].get_text()
```

```python
print(열량, 칼로리, 탄수화물, 설탕당, 단백질, 지방, 포화지방, 트랜스지방, 콜레스테롤, 나트륨)
```
