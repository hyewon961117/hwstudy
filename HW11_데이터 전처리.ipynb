# 값 변환 conversion

```python
df = pd.read_csv('./auto-mpg.csv', header = None)
df.columns = ['mpg', 'cylinders', 'displacement','horsepower','weight',
              'acceleration','model year', 'origin', 'name']

# mpg (mile per gallon)를 kpl(kilometer per liter)을 생성(mpg_to_kpl = 0.425)
mpg_to_kpl = 1.60934 / 3.78541

# mpg 열에 0.425를 곱해서 새로운 kpl 컬럼을 생성 (파생변수)
df['kpl'] = df['mpg'] * mpg_to_kpl

#kpl 열을 소수점 아래 둘째자리 반올림
df['kpl'] = df['kpl'].round(2)
df.head(3)
```

```python
# 각 열의 자료형 확인
df.dtypes
```

```python
# hores 열의 고유값
print(df['horsepower'].unique())

# ? 가 있기때문에 int로 바꾸면 에러가 뜰것임
```

```python
# 누락데이터('?')를 결측값으로 처리하기
# 누락데이터 ('?') 삭제
df['horsepower'].replace('?', np.nan, inplace=True)
# print(df.isnull().sum())
df.dropna(subset=['horsepower'], axis=0, inplace=True) # 특정 열의 행을 다 지울것이다.
print(df.isnull().sum())
# df.info() 로 확인후 ['horsepower'] 데이터 타입을 float64으로 변경
df['horsepower'] = df['horsepower'].astype('float')
df.info()
```

```python
# origin 열의 고유값을 확인
df['origin'].unique()
```

```python
df['origin'].replace({1:'USA',2:'EU',3:'JAPAN'}, inplace=True)
df['origin'].unique()
```

```python
# origin 열의 문자열 자료형을 범주형 자료로 변환
df['origin'] = df['origin'].astype('category')
print(df['origin'].dtypes)
```

# Horsepower를 3개의 등급으로 나누기

```python
#np.hisogram 함수로 3개의 bin으로 나누는 경계 값의 리스트 구하기
count, bin_dividers = np.histogram(df['horsepower'], bins=3)
print(bin_dividers)
```

```python
# 3개의 bins에 이름 지정
bin_name = ['저출력','보통출력','고출력']
```

```python
# cut 함수로 각 데이터를 3개의 bin에 할당
df['hp_bin'] = pd.cut(x=df['horsepower'],
                      bins = bin_dividers, # 경계값 리스트
                      labels = bin_name, # 해당구간 이름
                      include_lowest = True) # 첫 경계값을 포함
print(df[['horsepower','hp_bin']].head(10))
```

```python
# hp_bin 열의 범주형 데이터를 더미변수로 변환 (원핫인코딩)
horespower_dummies = pd.get_dummies(df['hp_bin'])
horespower_dummies.head(10)
```

```python
# 원핫인코딩을 다른 패키지로 사용 가능
# sklearn(사이킷런) 라이브러리 불러오기
# 주피터노트북 사용시 설치 필요
from sklearn import preprocessing
# 전처리를 위한 encoder 객체 만들기
label_encoder = preprocessing.LabelEncoder()
onehot_encoder = preprocessing.OneHotEncoder()
```

```python
#laber encoder로 문자열 범주를 숫자형 범주로 변환
onehot_labeled = label_encoder.fit_transform(df['hp_bin'].head(15))
print(onehot_labeled)
print(type(onehot_labeled))
```

```python
# 2차원 행렬형태
onehot_reshaped = onehot_labeled.reshape(len(onehot_labeled),1) # 열이 1개인 벡터 형태로 변환
print(onehot_reshaped)
print(type(onehot_reshaped))
```

```python
# 희소행렬로 변환 (1의 위치를 반환해주는 행렬 형태 만들기)
# (행과 열 튜플형태)에 값이 1이다.
onehot_fitted = onehot_encoder.fit_transform(onehot_reshaped)
print(onehot_fitted)
print(type(onehot_fitted))
```

```python
# 1. age 결측값을 평균으로 대체
# 2. embarked  최빈값으로 대체
# 3. embark_town 결측값이 있는 행 삭제
# 4. deck 삭제
# 5. titanic['adult/child'] 컬럼을 만들고 20살 이상이면 adult 이하면 child (apply, lambda 활용)
# 6. bins = [1, 20, 30, 50, 70, 100]
#    labels = ['미성년자', '청년', '중년', '장년', '노년']
# age를 기준으로 'age_cut' 컬럼을 생성후 두 컬럼을 출력하기
# age, age_cut
# 22   청년
# 38   중년
# pd.cut()
```

```python
# 1. age 결측값을 평균으로 대체
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
titanic

# 답
titanic.age.fillna(titanic.age.mean(), inplace=True)
```

```python
# 2. embarked  최빈값으로 대체
titanic['embarked'].fillna(titanic['embarked'].value_counts()[0],inplace=True)
titanic

# 답
titanic.embarked.value_counts()
titanic.embarked.fillna('S', inplace=True)
```

```python
titanic.info()
```

```python
# 3. embark_town 결측값이 있는 행 삭제
titanic = titanic[titanic['embark_town'].isnull()!=True]
# df.dropna(subset=['embark_town'], axis=0, inplace=True)
titanic.info()

# 답
titanic.dropna(subset=['embark_town'], inplace=True)
```

```python
# 4. deck 삭제
titanic.drop('deck', axis=1, inplace=True)
titanic.head()

# 답
titanic.drop(columns=['deck'], inplace=True)
titanic.shape
```

```python
# 5. titanic['adult/child'] 컬럼을 만들고 20살 이상이면 adult 이하면 child (apply, lambda 활용)
titanic['adult/child'] = titanic['age'].apply(lambda x : 'adult' if x>=20 else 'child')
titanic.head()

# 답
titanic['adult/child'] = titanic.age.apply(lambda x : 'adult' if x>=20 else 'child')
titanic.tail()
```

```python
# 6. bins = [1, 20, 30, 50, 70, 100]
#    labels = ['미성년자', '청년', '중년', '장년', '노년']
# age를 기준으로 'age_cut' 컬럼을 생성후 두 컬럼을 출력하기
# age, age_cut
# 22   청년
# 38   중년
# pd.cut()

bins = [1, 20, 30, 50, 70, 100]
labels = ['미성년자', '청년', '중년', '장년', '노년']
titanic['age_cut'] = pd.cut(titanic.age, bins, labels=labels)
titanic[['age','age_cut']]
```

# groupby

```python
# df.groupby(그룹기준컬럼).통계적용컬럼.통계함수
# 통계함수 # pivot_table에서는 함수 문자처리 했었음 
# .count() : 누락값을 제외한 데이터수
# .size() : 누락값을 포함한 데이터수
# .mean(), .sum(), .std(), . min(), .max()
```

```python
df_titanic.head()
```

```python
# 객실등급(pclass) 별 승선자수 결과 집계
df_titanic1 = df_titanic.groupby('pclass').survived.count().to_frame() # 데이터 프레임으로 바꿈
df_titanic1
```

```python
# 객실등급(pclass) 별 생존자수 결과 집계
df_titanic2 = df_titanic.groupby('pclass').survived.sum().to_frame() # 데이터 프레임으로 바꿈
df_titanic2
```

```python
# 객실등급(pclass) 별 생존비율 결과 집계
df_titanic3 = df_titanic.groupby('pclass').survived.mean().to_frame() # 데이터 프레임으로 바꿈
df_titanic3
```

```python
df_titanic4 = pd.concat([df_titanic1,df_titanic2,df_titanic3], axis=1)
df_titanic4.columns = ['승선자수', '생존자수', '생존비율']
df_titanic4
```

```python
# 성별 생존통계
df_titanic5 = df_titanic.groupby(['sex','pclass']).survived.count().to_frame()
df_titanic6 = df_titanic.groupby(['sex','pclass']).survived.sum().to_frame()
df_titanic7 = df_titanic.groupby(['sex','pclass']).survived.mean().to_frame()
df_titanic8 = pd.concat([df_titanic5,df_titanic6,df_titanic7],axis=1)
df_titanic8
df_titanic8
```
