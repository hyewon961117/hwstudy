## 시도별 전출입 인구수 데이터 시각화

```python
import matplotlib.pyplot as plt
df = pd.read_excel("시도별 전출입 인구수.xlsx", engine = "openpyxl", header = 0)
df.head(3) # default 5개, 숫자를 넣으면 개수만큼 출력
```

```python
df.info()
```

```python
df.describe()
```

```python
# 누락값(NAN)에 값을 채우는 함수
df = df.fillna(method ='ffill')  # ffill은 forward로 fill을 함 (전값으로 채워라)
# ffill() 과 pad() 같은 기능
# bfill() 뒷값으로 채워라
# 숫자나 문자열 넣는 경우 해당 값으로 고정하고 채움
# 딕셔너리 형태로 넣을 경우 각 열마다 채울 값 적어서 다르게 처리 할 수 있음
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask = (df['전출지별']=='서울특별시') & (df['전입지별']!='서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(["전출지별"],axis =1) # 전출지별 첫번째 열 삭제
df_seoul.rename({"전입지별":"전입지"},axis=1, inplace=True) # 열이름 전입지별에서 전입지로 변경
# df_seoul.rename(columns= {"전입지별":"전입지"})
df_seoul.set_index("전입지", inplace = True) # 전입지 열을 인덱스로 지정
df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc["경기도"]
sr_one
```

## 남북한 전력량으로 secondary_y를 그리기

```python
df = pd.read_excel('./남북한발전전력량.xlsx', engine='openpyxl', convert_float=True)
```

```python
# 북한 데이터만 가져오기
df = df.loc[5:9]
df.drop('전력량 (억㎾h)', axis='columns', inplace = True)
df.head()
```

```python
# 인덱스를 발전 전력별로 변경
df.set_index('발전 전력별',inplace = True)
```

```python
df = df.T
df.head()
```

```python
# 증감율(변동률) 계산
df = df.rename(columns={'합계':'총발전량'})
df.head()
```

```python
df['총발전량(전년)'] = df['총발전량'].shift(1) # 앞의 데이터
df.head()
```

```python
# 파생변수 만들기 - 데이터 내에 새로운 변수
df['증감율'] = ((df['총발전량'] / df['총발전량(전년)'])-1)*100
df.head()
```

### 2축 그래프

```python
# 2축 그래프 그리기
ax1 = df[['수력','화력']].plot(kind='bar', figsize=(20,10), width=0.7, stacked=True)
ax2 = ax1.twinx() # 여러 개의 그래프를 한 화면에 나타내기
ax2.plot(df.index, df.증감율, ls='--', marker='o', markersize=20,
         color='green', label='전년대비 증감율(%)') 
#ls : line style (라인 스타일)
ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('연도', size=20)
ax1.set_ylabel('발전량(억 Kwh)')
ax2.set_ylabel('전년대비 증감율')
plt.title('북한 전력 발전량(1990~2016)', size=25)
ax1.legend(loc='upper left')
plt.show()
```

### regplot

```python
sns.set_style('darkgrid')
# 스타일 테마 기본 5가지 : darkgrid, whitegrid, dark, white, ticks
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# 그래프 그리기 - 선형회귀선 표시
sns.regplot(x='age', y='fare', data=titanic, ax=ax1) # 회귀선(분석) - regression

# 그래프 그리기 - 선형회귀선 미표시
sns.regplot(x='age', y='fare', data=titanic, ax=ax2, fit_reg=False) #default = True
plt.show()
```

### scatter

```python
# seaborn의 scatter
sns.set_style('whitegrid')
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# 이산형 변수의 분포 - 데이터의 분산 미고려
sns.stripplot(x='class', y='age', data=titanic, ax=ax1)

# 이산형 변수의 분포 - 데이터의 분산 고려 (중복X)
sns.swarmplot(x='class', y='age',data=titanic, ax=ax2)

# 차트의 제목
ax1.set_title('Stripplot')
ax2.set_title('Swarmplot')
plt.show()
```

### bar plot

```python
# bar plot
sns.set_style('whitegrid')
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# x축과 y축에 변수를 할당
sns.barplot(x='sex', y='survived', data=titanic, ax=ax1)

# x축과 y축에 변수를 할당 hue 옵션 class
sns.barplot(x='sex', y='survived', hue='class', data=titanic, ax=ax2)

# x축과 y축에 변수를 할당 hue 옵션 class 누적으로 표시
sns.barplot(x='sex', y='survived', hue='class', dodge=False, data=titanic, ax=ax3)

ax1.set_title('titanic survived - sex') # 여성들이 생존률이 높음
ax2.set_title('titanic survived - sex/classs') # 부자들이 생존률이 높음
ax3.set_title('titanic survived - sex/classs(stacked)') # 누적으로 보임

plt.show()
```

### count plot

```python
# count plot - 빈도 그래프
sns.set_style('whitegrid')
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# 기본값에서
# palette = 색 구성을 다르게 하기
# seaborn 팔레트 검색해서 찾아보기
sns.countplot(x='class', palette='Set1', data=titanic, ax=ax1)

#class, who 컬럼의 값별로 count 
# hue 옵션으로 who를 추가 (여자, 남자, 아이)
sns.countplot(x='class', palette='Set2', hue='who', data=titanic, ax=ax2)

# class, who 컬럼의 값별로 count + 누적 출력
# 축 방향으로 hue 변수를 분리하지 않고 위로 쌓아 올리는 누적그래프
sns.countplot(x='class', palette='dark', hue='who', dodge=False, data=titanic, ax=ax3)


ax1.set_title('titanic calss') 

plt.show()
```

### box plot

```python
# box plot
sns.set_style('whitegrid')
fig = plt.figure(figsize = (15,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# ax1 boxplot - 기본값
sns.boxplot(x='alive', y='age', data=titanic, ax=ax1)

# ax2 boxplot - hue = sex
sns.boxplot(x='alive', y='age', hue='sex', data=titanic, ax=ax2)

#ax3 바이올린그래프 - 기본값
sns.violinplot(x='alive', y='age', data=titanic, ax=ax3)

#ax4 바이올린그래프 - hue = sex
sns.violinplot(x='alive', y='age', hue='sex', data=titanic, ax=ax4)

# 바이올린 그래프
# 1. 중심선을 따라 대칭인 kde plot
# 2. 가운데 보이는 흰색 점은 중앙값
# 3. 두꺼운 선은 사분위 범위를, 얇은 선은 (95%)신뢰구간을 의미

plt.show()
```

### join plot

```python
# join plot : 산점도 + 히스토그램
sns.set_style('whitegrid')

# 조인트그래프 - 기본값
j1 = sns.jointplot(x='fare', y='age', data=titanic)

# 조인트그래프 - 회귀선
j2 = sns.jointplot(x='fare', y='age', kind='reg', data=titanic)

# 조인트 그래프 - 육각 산점도
# kind에 hexbin을 넣으면 육각 산점도를 그릴 수 있다.
j3 = sns.jointplot(x='fare', y='age', kind='hex', data=titanic)

# 조인트 그래프 - 커널 밀집 그래프
j4 = sns.jointplot(x='fare', y='age', kind='kde', data=titanic)

# 차트제목
j1.fig.suptitle('titanic fare - scatter', size=15)
j2.fig.suptitle('titanic fare - reg', size=15)
j3.fig.suptitle('titanic fare - hex', size=15)
j4.fig.suptitle('titanic fare - kde', size=15)
plt.show()
```

```python
# 경기도 인구변화 데이터를 불러와서 데이터 프레임으로 변환
df2 = pd.read_excel('./경기도인구데이터.xlsx', index_col='구분', engine='openpyxl')
df2.columns = df2.columns.map(str)
df2.head()
```

```python
# 경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기
import json
geo_path = './경기도행정구역경계.json'
try : 
    geo_data = json.load(open(geo_path, encoding='utf-8'))
except:
    ged_data = json.load(open(geo_path, encoding='utf-sig'))
```

```python
# 경기도 지도 만들기
g_map = folium.Map(location=[37.5502, 126.982],
                  tiles='Stamen Terrain', zoom_start=9)
```

```python
# 출력할 연도를 선택(2007~2017)
year = '2017'
# 단계구분도(Choropleth) 단계구분도 표시하기
folium.Choropleth(geo_data=geo_data, data=df2[year],
                 columns=[df2.index, df2[year]],
                 fill_color='YlOrRd',
                 fill_opacity=0.7, line_opacity = 0.3, #opacity : 투명도
                 threshold_scale=[10000,100000,300000,500000,7000000],
                 key_on = 'feature.properties.name').add_to(g_map)
```

```python
# 지도를 Html로 저장
g_map.save('./gyonggi_population' + year + '.html')
```
