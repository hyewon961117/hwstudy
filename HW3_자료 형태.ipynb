리스트(튜플) 통계값 구하기
* 항목수 : len(리스트명)
* 합계 : sum(리스트명)
* 최소값 : min(리스트명)
* 최대값 : max(리스트명)
* 항목의개수 : 리스트명.count(항목)

```python
lst = [1,2,3,4,5,3,4,5]
print(len(lst)) # 항목수
print(sum(lst)) # 합계
print(min(lst)) # 최소값
print(max(lst)) # 최대값
print(lst.count(3)) # 항목의 개수
print(sum(lst)/len(lst)) # 평균
```

```python
tp = [1,2,3,4,5,3,4,5]
print(len(tp)) # 항목수
print(sum(tp)) # 합계
print(min(tp)) # 최소값
print(max(tp)) # 최대값
print(tp.count(3)) # 항목의 개수
print(sum(tp)/len(tp)) # 평균

# 튜플은 수정, 삭제만 안될 뿐 리스트와 똑같이 계산됨

# 코랩에서 리스트 명 한번에 바꾸고 싶을 때 리스트 드래그하고 ctrl+d 누르면 됨
```

딕셔너리를 만드는 방법 3가지

```python
# 1. 딕셔너리명 = dict(키1=값1, 키2=값2, 키3=깂3 ....)
# 주의 키 1.2.3..에 ''("")을 쓰지 않음
menu1 = dict(김밥=3000, 떡볶이=2000, 어묵=3500, 튀김=2800)
print(menu1)
```

```python
# 2. 딕셔너리명 = dict(zip(key리스트, value리스트))
key_list = ['김밥', '떡볶이', '어묵', '튀김']
value_list = [3000, 2000, 3500, 2800]
menu2=dict(zip(key_list, value_list))
print(menu2)
```

```python
# 3. 딕셔너리명 = dict([(키1, 값1),(키2, 값2)...])
menu3 = dict([('김밥',3000),('떡볶이',2000),
              ('어묵',3500),('튀김',2800)])
print(menu3)
```

```python
# setdefault(key, [default])
# key가 딕셔너리에 있으면 해당값을 돌려줍니다.
# 그렇지 않으면 default값을 갖는 key를 삽입한 후 default를 돌려줌
#default의 값에 'None'값이 생성됨

score = {'kor':100, 'eng':90, 'math':80}
print(score.setdefault('kor'))

print(score.setdefault('music')) # key값이 없어도 None이라는 값을 넣어 생성됨
print(score)
```

```python
score = {'kor':100, 'eng':90, 'math':80}

print(score.setdefault('music', 100)) # key값이 없다면 값을 넣어 생성됨
print(score)

print(score.setdefault('kor',44)) # key 값이 기존에 있으면 value값이 변하지 않음
print(score)

score['kor'] = 44 # key 값이 기존에 있으면 value값을 변경하는 법
print(score)
```

```python
wc = {'국민':0, '존경':0}

print(wc.get('국민')) # 키에 있는 값이면 해당 값 반환
print(wc)

print(wc.get('사랑')) # 키에 없는 값이면 None을 반환, 다만 원소에 생성X
print(wc)

print(wc.get('국민',1)) # 키에 있는 값이면 값을 변경하지않고 키에 있는 값을 반환
print(wc)

print(wc.get('사랑',0)) # 키에 없는 값이면 0을 반환, 다만 원소에 생성X
print(wc)

print(wc.setdefault('사랑',0)) # 키에 없는 값이면 추가 하고 값에 0 할당
print(wc)

print(wc.setdefault('사랑',1)) # 키에 있는 값이면 값을 변경하지않고 키에 있는 값을 반환
print(wc)
```

```python
#아래 코드는 비트코인의 가격 정보를 딕셔너리로 가져오는 코드이다.
import requests
btc = requests.get("https://api.bithumb.com/public/ticker/").json()['data']
print(btc)

# btc 딕셔너리 안에는 시가, 종가, 최고가, 최저가 등이 저장되어 있다.
# 최고가와 최저가의 차이를 변동폭으로 정의할 때 
# (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 
# 그렇지 않은 경우 "하락장" 문자열을 출력하라.

price=int(btc['max_price'])-int(btc['min_price'])
print(price)

score=int(btc['opening_price']) + price

if score > int(btc['max_price']):
  result = "상승장"
elif score < int(btc['min_price']):
  result = "하락장"
else :
  result = "같다"

print(result)
```

```python
# 답
#아래 코드는 비트코인의 가격 정보를 딕셔너리로 가져오는 코드이다.
import requests
btc = requests.get("https://api.bithumb.com/public/ticker/").json()['data']

# btc 딕셔너리 안에는 시가, 종가, 최고가, 최저가 등이 저장되어 있다.
# 최고가와 최저가의 차이를 변동폭으로 정의할 때 
# (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 
# 그렇지 않은 경우 "하락장" 문자열을 출력하라.

시가 = float(btc['opening_price'])
최고가 = float(btc['max_price'])
최저가 = float(btc['min_price'])
변동폭 = 최고가 - 최저가

if 시가 + 변동폭 > 최고가:
  print("상승장")
else :
  print("하락장")
```
