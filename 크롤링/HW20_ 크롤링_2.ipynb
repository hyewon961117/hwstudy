# 연습사이트 1

```python
import requests
from bs4 import BeautifulSoup
url = "https://crawlingstudy-dd3c9.web.app/01/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.select_one("p#cook").get_text()
```

```python
soup.select_one("p#cook").string
```

```python
soup.select_one("p#cook").text
```

```python
dic={}
lst=[]
keys=[]

for th in soup.select("th"):
    keys.append(th.string)

for tr in soup.find('tbody').select("tr"):
    values=[]
    for td in tr.select("td"):
        values.append(td.string)
    lst.append(dict(zip(keys,values)))

lst
```

```python
# 답
# [{'이름': '이몽룡', '나이': '34'}, {'이름': '홍길동', '나이': '23'}]

key = []
value = []

for element in soup.find('table').find_all('th'):
    key.append(element.text)
    
for element in soup.find('table').find('tbody').find_all('tr'):
    temp = []
    for td_element in element.find_all('td'):
        temp.append(td_element.text)
    value.append(dict(zip(key, temp)))

value
```

```python
import requests
from bs4 import BeautifulSoup
url = "https://crawlingstudy-dd3c9.web.app/03/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# [['써니전자', '5,000'], ['삼성전자', '55,200'], ['안랩', '81,000'], ..., ['SK텔레콤', '234,000']]
popular = []

soup.select(".lst_pop")
soup.select('#popularItemList>li') # 리스트로 담김

for item in soup.select('#popularItemList>li'):
#     print(item)
    popular.append([item.select_one('a').text, item.select_one('span').text])

print(popular)
```

```python
for i in range(1,3):
    print('PAGE ====================', i)
    url = f'https://www.musinsa.com/categories/item/001?d_cat_cd=001&brand=&list_kind=small&sort=pop_category&sub_sort=&page={i}&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&kids=&color=&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure='
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
#     print(soup)
    for item in soup.select("ul#searchList>li"):
        print('상품명:', item.select_one(".article_info p.list_info a").attrs['title'])
        print('브랜드:', item.select_one(".article_info p.item_title").text.strip())
        price =""
        
        # 원래 가격과 세일가격 2개가 있는 경우가 있기 때문에 길이를 1또는 else로 처리
        if len(item.select_one('.article_info p.price').text.strip().split(" ")) == 1:
            price = item.select_one('.article_info p.price').text.strip().split(" ")[0]
        else:
            price = item.select_one('.article_info p.price').text.strip().split(" ")[16]
        print('가격: ', price)
        print("")
```

```python
for i in range(1,3):
    print('PAGE ====================', i)
    url = f'https://www.musinsa.com/categories/item/001?d_cat_cd=001&brand=&list_kind=small&sort=pop_category&sub_sort=&page={i}&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&kids=&color=&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure='
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
#     print(soup)
    for item in soup.select("ul#searchList>li"):
        print('상품명:', item.select_one(".article_info p.list_info a").attrs['title'])
        print('브랜드:', item.select_one(".article_info p.item_title").text.strip())
        price =""
        
        # 가격 정규표현식 이용 방법
        if len(re.findall('[0-9,]*원', item.select_one('.article_info p.price').text)) == 1:
            price = re.findall('[0-9,]*원', item.select_one('.article_info p.price').text)[0]
        else:
            price = re.findall('[0-9,]*원', item.select_one('.article_info p.price').text)[1]
        print('가격: ', price)
        print("")
```
