```python
driver = webdriver.Chrome() # 팝업창 끄면 동작 안함
url = "https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page=1"
driver.get(url)
time.sleep(2)
```

```python
#soup
soup = BeautifulSoup(driver.page_source, 'html.parser')
soup
```

```python
trs = soup.select(".aos-init")
len(trs)
```

```python
tr = trs[0]
```

```python
# 순위 
rank = tr.select_one('.rank').get_text().strip()

# 카테고리
category = tr.select_one('.category').get_text().strip()[1:-1]

# 채널명
channel = tr.select_one('.subject a').get_text().strip()

# 구독자 조회수 비디오개수
subscriber = tr.select_one('.subscriber_cnt').get_text().strip()
view = tr.select_one('.view_cnt').get_text().strip()
video = tr.select_one('.video_cnt').get_text().strip()
```

```python
subscriber, view, video
```

```python
# 첫 페이지 크롤링
lines = []
for tr in trs:
    rank = tr.select_one('.rank').get_text().strip()
    category = tr.select_one('.category').get_text().strip()[1:-1]
    channel = tr.select_one('.subject a').get_text().strip()
    subscriber = tr.select_one('.subscriber_cnt').get_text().strip()
    view = tr.select_one('.view_cnt').get_text().strip()
    video = tr.select_one('.video_cnt').get_text().strip()
    lines.append([rank, category, channel, subscriber, view, video])
```

```python
df = pd.DataFrame(lines, columns=['순위', '카테고리', '채널', '구독자수', '조회수', '비디오수'])
df
```

```python
# 숫자로 바꿔주는 함수 만들기
def convert_unit(s):
    s = s.replace("억","").replace("개","").replace(",","").replace("만","0000")
    return f'{int(s):,d}' # 숫자에 쉼표를 넣어서 반환
```

```python
for column in df.columns[3:]:
    df[column]=df[column].apply(convert_unit)
df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm.notebook import tqdm

import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome() # 팝업창 끄면 동작 안함
url = "https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page=1"
driver.get(url)
```

```python
driver.find_elements('css selector','tr')[1].text.split()
```

```python
lines = []
for page in tqdm(range(1,11)):
    driver = webdriver.Chrome()
    url = "https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page=" + str(page)
    driver.get(url)
    time.sleep(2)
    trs = driver.find_elements('css selector','tr.aos-init')
    
    for tr in trs:
        rank = tr.find_element('css selector','.rank').text.strip()
        category = tr.find_element('css selector','.category').text.strip()[1:-1]
        channel = tr.find_element('css selector','.subject a').text.strip()
        subscriber = tr.find_element('css selector','.subscriber_cnt').text.strip()
        view = tr.find_element('css selector','.view_cnt').text.strip()
        video = tr.find_element('css selector','.video_cnt').text.strip()
        lines.append([rank, category, channel, subscriber, view, video])
    driver.close()
```

```python
df = pd.DataFrame(lines, columns=['순위', '카테고리', '채널', '구독자수', '조회수', '비디오수'])
df
```
