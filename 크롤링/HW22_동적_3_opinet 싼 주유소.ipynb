# Opinet 싼 주유소 찾기 - 지역별

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm.notebook import tqdm

import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
url = "https://www.opinet.co.kr/user/main/mainView.do"
driver.get(url)
time.sleep(2)
```

```python
# 팝업창 싼 주유소 찾기 클릭
driver.find_element(By.CSS_SELECTOR, '.ic_m1').click()
```

```python
# 지역 - 서울 copy Xpath
metro = driver.find_element(By.XPATH, '//*[@id="SIDO_NM0"]')
metro.send_keys("서울")
time.sleep(1)
```

```python
metro.text.split()
```

```python
# 시/군/구 XPATH
sigungu = driver.find_element(By.XPATH,'//*[@id="SIGUNGU_NM0"]')
gu_list = sigungu.find_elements(By.CSS_SELECTOR, 'option')
```

```python
gu_name = []
for gu in gu_list:
    gu_name.append(gu.text)
gu_name
```

```python
gu_name = [gu.text for gu in gu_list]

del gu_name[0]

gu_name
```

```python
# 엑셀 저장 버튼
driver.find_element(By.XPATH,'//*[@id="glopopd_excel"]/span').click()
```

```python
# 25개 자치구에 대해서 주유소 정보 엑셀 저장
for gu in gu_name:
    sigungu = driver.find_element(By.XPATH,'//*[@id="SIGUNGU_NM0"]')
    time.sleep(3)
    sigungu.send_keys(gu)
    time.sleep(5)
    driver.find_element(By.XPATH,'//*[@id="glopopd_excel"]/span').click()
```

# 서울시 구별 주유소 데이터 분석 및 시각화

```python
import numpy as np
import pandas as pd

import requests
from urllib.parse import quote

import folium, json

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./서울_주유소(230418).csv")
df.head()
```

```python
# 구별 집계 분석
gu_mm = df.groupby('구')["가격"].agg(['mean', 'min', 'max'])
gu_mm
```

```python
# 해당구에서 최대값 최소값만 구하기
# 예 강남구
_, mi, ma = gu_mm.loc['강남구'] # mean은 안쓸거라 변수명 _로 익명처리
mi, ma = int(mi), int(ma)
mi, ma
```

```python
# 컬럼을 하나 추가 (최대값인지 최소값인지 결정해서 넣을 예정)
df['mm'] = np.zeros(df.shape[0], int)
df.head()
```

```python
# mm 컬럼에 최소값이 같으면 1을 넣고 최대값과 같으면 2를 넣을 것

for gu in gu_mm.index:
    _, mi, ma = gu_mm.loc[gu]
    mi, ma = int(mi), int(ma)
    for i in df.index:
        if df['가격'][i] == mi and df['구'][i] == gu:
            df['mm'][i] = 1
        if df['가격'][i] == ma and df['구'][i] == gu:
            df['mm'][i] = 2
```

```python
df.head()
```

```python
# mm 컬럼에 데이터가 있는 행만 추출
df1 = df[df.mm != 0]
df1.head()
```

```python
df1.shape
```

## 지도에 시각화

```python
# KAKAO API로 위, 경도값을 추출
with open('C:/workspace/APIkey/카카오API.txt') as file:
    kakao_key = file.read()
```

```python
lat_list = []
lng_list = []

for i in df1.index :
    search_url = "https://dapi.kakao.com/v2/local/search/address.json"
    url = f'{search_url}?query={quote(df1["주소"][i])}'
    result = requests.get(url, 
                          headers={"Authorization": f'KakaoAK {kakao_key}'}).json()
    try : 
        lng_list.append(float(result['documents'][0]['x']))
        lat_list.append(float(result['documents'][0]['y']))
    except : 
        print(df1['상호'][i])
```

```python
df1['위도'] = lat_list
df1['경도'] = lng_list
df1.head()
```

```python
# 지도위에 주유소 표시
filename = "./gasData/skorea_municipalities_geo_simple.json"
geo_str = json.load(open(filename, encoding='utf-8'))
geo_str
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=12)
for i in df1.index:
    folium.Marker(location = [df1.위도[i], df1.경도[i]],
                 popup=folium.Popup(df1.주소[i], max_width=200),
                 tooltip=f'{df1.상호[i]}<br>휘발유:{df1.가격[i]:,d}원',
                 icon = folium.Icon(color="green" if df.mm[i] == 1 else 'red',
                                   icon = 'glyphicon glyphicon-thumbs-up' if df.mm[i] == 1 else 'glyphicon glyphicon-thumbs-down')).add_to(map)
map
```
