```python
lines = []

for i in range(1,5):
    header = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
    url = f"https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230417&hh=15&rtm=Y&pg={i}"
    response = requests.get(url, headers=header)
    soup = BeautifulSoup(response.text,'html.parser')

    for tr in soup.select("tr.list"):
        rank = tr.select_one(".number").get_text().split()[0]
        title = tr.select_one(".title").get_text().strip()
        artist = tr.select_one(".artist").get_text().strip()
        album = tr.select_one(".albumtitle").get_text().strip()
        
        lines.append([rank, title, artist, album])
lines
```

```python
df = pd.DataFrame(lines, columns=["순위","제목","가수","앨범"])
df
```

```python
lines = []
for page in range(1,5):
    base_url = "https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230417&hh=15&rtm=Y&pg="
    url = base_url + str(page)
    header = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}

    req = requests.get(url, headers = header)
    soup = BeautifulSoup(req.text,'html.parser')
    
    trs = soup.select('#body-content > div.newest-list > div > table > tbody > tr')
    
    for tr in trs:
        try:
            rank = tr.select_one(".number").get_text().split("\n")[0]
        except:
            rank = 0
        title = tr.select_one(".title.ellipsis").get_text().strip()
        artist = tr.select_one(".artist.ellipsis").get_text().strip()
        album = tr.select_one(".albumtitle.ellipsis").get_text().strip()
        
        lines.append([rank, title, artist, album])
lines
```

```python
df = pd.DataFrame(lines, columns=["순위","제목","가수","앨범"])
df
```
