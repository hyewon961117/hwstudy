# 실제 사이트 _ 다음 영화 순위 정보

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
url = "https://search.daum.net/search?nil_suggest=btn&w=tot&DA=SBC&q=%EC%98%81%ED%99%94+%EC%88%9C%EC%9C%84"

import requests
from bs4 import BeautifulSoup

response = requests.get(url, headers = headers)
soup = BeautifulSoup(response.text,'html.parser')
```

```python
soup
```

```python
movieInfoList = soup.find("ol",attrs={'class':'movie_list'}).find_all('li')
movieInfoList
```

```python
movieInfo = movieInfoList[0]
movieInfo
```

```python
movieRank = movieInfo.find("span",{"class":"img_number"}).get_text()
movieRank
```

```python
movieTitle = movieInfo.find('a', attrs={'class':'tit_main'}).get_text()
movieTitle.strip().replace(" ","")
```

```python
# 점수
movieScore = movieInfo.find('em', attrs={'class':'rate'}).get_text()
movieScore
```

```python
# 평가참여자수
movieScoreCnt = movieInfo.find('a', attrs={'class':'link_count'}).get_text()
movieScoreCnt = int(movieScoreCnt.replace('명 참여',""))
movieScoreCnt
```

```python
# 예매율
movieTicketSales = movieInfo.find('dd', attrs={'class':'cont'}).get_text()
movieTicketSales
```

```python
movieOpenDate = movieInfo.find_all('dd', attrs={'class':'cont'})[1].get_text()
movieOpenDate.strip()
```

```python
movieList=[]
for item in movieInfoList:
    movieRank = int(item.find("span",{"class":"img_number"}).text.replace('위',""))
    movieTitle = item.find("a",{"class":"tit_main"}).text.strip().replace(" ","")
    if item.find("em",{"class":"rate"}).text == "":
        movieScore = 0
        movieScoreCnt = 0
    else :
        movieScore = float(item.find("em",{"class":"rate"}).text)
        movieScoreCnt = int(item.find("a",{"class":"link_count"}).text.replace('명 참여',"").replace(",",""))
    if len(item.find_all("dd",{"class":"cont"}))==1:
        movieTicketSales = item.find_all("dd",{"class":"cont"})[0].text
        movieOpenDate = 0
        moviePeople = 0
    elif len(item.find_all("dd",{"class":"cont"}))==2:
        movieTicketSales = item.find_all("dd",{"class":"cont"})[0].text
        movieOpenDate = item.find_all("dd",{"class":"cont"})[1].text.strip()
        moviePeople = 0
    else :
        movieTicketSales = item.find_all("dd",{"class":"cont"})[0].text
        movieOpenDate = item.find_all("dd",{"class":"cont"})[1].text.strip()
        moviePeople = int(item.find_all("dd",{"class":"cont"})[2].text.replace("명","").replace(",",""))
    
    movieList.append({"영화순위": movieRank,
                     "영화제목" : movieTitle,
                     "점수" : movieScore,
                     "평가참여자수" : movieScoreCnt,
                     "예매율" : movieTicketSales,
                     "개봉일" : movieOpenDate,
                     "누적" : moviePeople})
movieList
```

## 강사님 방법

```python

for movieInfo in movieInfoList:
    movieRank = movieInfo.find('span', attrs={'class':f'img_number'}) 
    movieTitle = movieInfo.find('a', attrs={'class':'tit_main'})
    movieScore = movieInfo.find('em', attrs={'class':'rate'})
    movieScoreCnt = movieInfo.find('a', attrs={'class':'link_count'})
    ticketSalesAndOpenDate = movieInfo.find_all('dd', attrs={'class':'cont'})
    if len(ticketSalesAndOpenDate) > 1:
        movieTicketSales = ticketSalesAndOpenDate[0]
        movieOpenDate = ticketSalesAndOpenDate[1]
    else: # 개봉일자가 없을 경우
        movieTicketSales = ticketSalesAndOpenDate[0]
        
    # 개봉일자 다름,,
    print(f'순위 : {movieRank.get_text() if movieRank else "X"}')
    print(f'제목 : {movieTitle.get_text() if movieTitle else "X"}')
    print(f'점수 : {movieScore.get_text() if movieScore else "X"}')
    print(f'평가참여수 : {movieScoreCnt.get_text() if movieScoreCnt else "X"}')
    print(f'예매율 : {movieTicketSales.get_text() if movieTicketSales else "X"}')
    print(f'개봉일자 : {movieOpenDate.get_text().strip() if movieOpenDate else "X"}')
    print("-------------------------------")
```
