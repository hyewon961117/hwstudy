```python
a=[1,2,3]
try :
  x = int(input())
  y = 100 / x
  print(y, a[x])
except ValueError :
  print('입력값이 잘못되었습니다.')
except ZeroDivisionError : 
  print('0으로 나눌 수 없음')
except IndexError:
  print('index 에러 발생')
```

```python
a=[1,2,3]
try :
  x = int(input())
  y = 100 / x
  print(y, a[x])
except Exception as e :
  print(e)
```

```python
# else - finally
try :
  x = int(input())
  y = 100/x
except Exception as e :
  print(e)
else:
  print(y) # 프린트문이 없으면 아무것도 반환하지 않음
```

```python
# else - finally
try :
  x = int(input())
  y = 100/x
except Exception as e :
  print(e)
else:
  print(y) # 프린트문이 없으면 아무것도 반환하지 않음
finally : # 예외발생여부와 상관없이 항상 실행
  print('코드실행종료')
```

#파일

```python
f = open('새파일.txt', "w")
for i in range(1, 11):
  data = "%d번째 줄입니다.\n"%i
  f.write(data)
f.close()
```

```python
f = open('새파일.txt','r')
line = f.readline()
print(line)
```

```python
f = open('새파일.txt','r')
while True:
  line = f.readline()
  if not line:
    break
  print(line)
f.close()
```

```python
f = open('새파일.txt', 'a')
for i in range(11,20):
  data = '%d번째 줄입니다.\n'%i
  f.write(data)
```

```python
with open('hello.txt','w',encoding = 'utf-8') as file:
  file.write('이글은 utf-8로 인코딩 되어있습니다')
  file.write('EUC-KR(CP 949)로 읽으면 글자가 깨집니다.')
```

```python
!cat hello.txt
```

```python
with open('hello.txt','r',encoding = 'utf-8') as file: # 인코딩으로 인한 에러 'cp949'
  text = file.read()
print(text)
```

```python
# 파이썬 객체 저장/로드
import pickle
person = {'name':'james', 'age':25, 'addr':'신촌'}
with open('person.pkl','wb') as file:
  pickle.dump(person, file) # person이라는 데이터를 저장할것이다.
```

```python
!cat person.pkl
```

```python
with open('person.pkl','rb') as file:
  james = pickle.load(file)
print(james)
```

```python
import joblib # 요즘엔 피클보다 더 많이 씀 (같은 함수)
joblib.dump(person, 'person.jl')
```

```python
!cat person.jl
```

```python
james1 = joblib.load('person.jl') # 피클보다는 코드가 좀 더 간단함
print(james1)
```
