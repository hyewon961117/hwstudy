```python
# 1. Character 클래스 만들기
# Character의 속성값에 Health 속성 ( 생성시 200 )
# Character 클래스 move() 메소드를 추가하고 메소드 사용시 health가 -10이 됨
# Character의 클래스 rest() 메소드를 추가하고 메소드 사용시 health +10이 됨
# 현재 health를 알 수 있는 checkHealth() method를 추가

# 2. Knight, Healer 클래스 만들기
# Knight와 Healer는 Character의 클래스를 상속
# Knight Move()를 하면 Health가 추가로 -5를 더 소모함
# Knight Attack()를 추가하고 '공격합니다.'라는 메소드를 작성

# Healer Mana라는 속성을 추가하고 생성시 100의 값을 주어짐
# Healer heal() 메소드를 추가하고 이 메소드는 Character들을 매개변수로 받음
# Healer heal() 메소드를 실행하면 Mana 가 -10이 되고
# 전달받은 Character 객체에 rest() 메소드가 실행되어 +10이 되도록 함
# Healer는 현재 마나 속성을 확인할 수 있는 Checkmana() 메소드를 추가
```

```python
class Character():
  def __init__(self):
    self.health = 200
  def Move(self):
    self.health -= 10
  def Rest(self):
    self.health += 10
  def checkHealth(self):
    print(f'현재 캐릭터의 체력은 : {self.health} 입니다')
```

```python
class Knight(Character): # 상속
  def __init__(self):
    super().__init__()
  def Move(self): # 오버라이드
    super().Move()
    self.health -= 5
  def Attack(self):
    print('공격합니다')
```

```python
class Healer(Character):
  def __init__(self):
    super().__init__()
    self.Mana=100
  def Heal(self,Character):
    Character.Rest()
    self.Mana -= 10
  def checkMana(self):
    print(f'현재 캐릭터의 마나는 : {self.Mana} 입니다')
```

```python
knight = Knight()
healer = Healer()
knight.checkHealth()
healer.checkHealth()
```

# 은행만들기

```python
# 속성 : 계좌번호 '123456' (ano) / owner : 이름 (문자열) / 잔액 : balance, 비공개 속성
# method : 저금(deposit), 출금(withdraw), 조회 출력이 가능하게
# 저금(deposit) : 1000만원 이상 입금할 수 없음 (통장 최대 1000만원 한도)
# 출금(withdraw) : 잔액이 0이면 출금이 안되도록
# 조회 출력이 가능하게
```

```python
class Account():
  def __init__(self, ano, owner, balance):
    self.ano = ano
    self.owner = owner
    self.balance = balance

  def deposit(self, amount):
    if amount+self.balance >= 10000000:
      print("1000만원 이상 입금할 수 없습니다.")
      return
    self.balance += amount

  def withdraw(self, amount):
    if self.balance - amount < 0:
      print('잔액이 부족합니다.')
      return
    self.balance -= amount


  def __str__(self): #__str__은 문자열로 표현한 것을 반환해주는 함수
    return f'계좌번호:{self.ano}, 소유주 : {self.owner}, 잔액 : {self.balance : 9,d}'
```

```python
acc = Account('12345', '제임스', 100000)
print(acc)
```

```python
acc.deposit(3000000)
print(acc)
```

```python
acc.deposit(7000000)
print(acc)
```

```python
acc.withdraw(100000)
print(acc)
```

```python
acc.withdraw(4000000)
print(acc)
```

```python
acc2 = Account('987654','마리아',3000000)
print(acc2)
```

```python
acc_list = [acc, acc2]
def check_duplicate(ano):
  for acc in acc_list : 
    if ano == acc.ano:
      return True
  return False
```

```python
# 사용자로부터 필요한 정보를 입력받아서 계좌 생성
def create_account():
  while True:
    s = input('계좌번호 성명 금액을 입력 > (한칸씩 띄어서 입력)').split()
    ano, owner = s[0], s[1]
    amount = int(s[2]) # 금액 형변환
    if not check_duplicate(ano): # 거짓이면 멈춰라 #계좌번호가 있으면 계속 돌아가게
      break
  acc = Account(ano, owner, amount)
  acc_list.append(acc)
```

```python
# 사용자로부터 필요한 정보를 입력받아서 계좌에 돈을 입금
def deposit_account():
  s=input('계좌번호 금액>').split()
  ano, amount = s[0], int(s[1])
  for acc in acc_list:
    if acc.ano == ano:
      acc.deposit(amount)
      return
```

```python
def withdraw_account():
  s=input('계좌번호 금액>').split()
  ano, amount = s[0], int(s[1])
  for acc in acc_list:
    if acc.ano == ano:
      acc.withdraw(amount)
      return
```

```python
# 화면
while True :
  menu = int(input('1:계좌생성, 2:계좌목록, 3:입금, 4:출금, 5:종료>'))
  if menu == 5:
    break
  if menu == 1:
    create_account()
  elif menu == 2:
    for account in acc_list:
      print(account)
  elif menu == 3:
    deposit_account()
  elif menu == 4:
    withdraw_account()
  print()
```
