```python
from statistics import mean, variance
from math import sqrt

# 산술평균

dataset = [2, 4 ,5 ,6, 1, 8]

def Avg(data):
  avg = mean(data) # mean 자체에 언패킹되게 구성되어있음
  return avg

print('산술평균 =',Avg(dataset))
```

```python
# 분산/표준편차
def var_sd(data):
  avg = Avg(data) # 함수 내에 만든 함수를 넣을 수 있음
  diff = [(d - avg)**2 for d in data]
  # diff = []
  # for d in data:
  #   c= (d - avg)**2
  #   diff.append(c)
  var = sum(diff) / (len(data)-1)
  sd = sqrt(var)
  return var, sd
```

```python
v, s = var_sd(dataset)
print("분산=",v)
print('표준편차',s)
```

```python
import random

# 동전 앞면과 뒷면 난수 확률분포 함수 정의
def coin(n):
  result = []
  for i in range(n):
    r = random.randint(0, 1)
    if (r == 1):
      result.append(1)
    else:
      result.append(0)
  return result
```

```python
# 몬테카를로 시물레이션
def montaCoin(n) :
  cnt = 0
  for i in range(n):
    cnt += coin(1)[0]
  result = cnt / n
  return result
```

```python
print(montaCoin(10))
print(montaCoin(30))
print(montaCoin(100))
print(montaCoin(1000))
print(montaCoin(10000))
print(montaCoin(100000)) # 많이 던지면 던질수록 확률 0.5에 수렴한다
```

```python
# 1. 세 수를 입력받아 큰수를 출력하는 함수를 만드시오.

def print_max(a,b,c):
  max_value = 0
  if a > max_value:
    max_value = a
  if b > max_value:
    max_value = b
  if c > max_value:
    max_value = c
  print(max_value)

print_max(1,2,8)
```

```python
# 2. print_5xn('string') 입력받아 한줄에 다섯글자씩만 출력하는 함수를 작성
# '아이엠어보이유알어걸' → 아이엠어보
#                           이유알어걸

def print_5xn(line):
  chek_num = int(len(line)/5)
  for x in range(chek_num+1):
    print(line[x*5 : x*5 + 5]) # 슬라이싱

print_5xn('아이엠어보이유알어')
```
